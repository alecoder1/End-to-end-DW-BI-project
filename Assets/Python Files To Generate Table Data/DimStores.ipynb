## Generating Random Values for DimStore.csv 

```python
# import python libraries
import pandas as pd
import csv
import random
from faker import Faker
```

```python
# Initialize faker
faker = Faker()

# input the number of rows that the csv file should have
num_rows = int(input("Enter the number of rows you want to generate in te csv file: "))

# input the name of the csv file with the (.csv extension)
csv_file = input("Enter the name of the csv file: ")

# Details of the excel file that has the lookup data, 
# File Path and Name, Sheet Name 
# and column names where the data is present

excel_file_path = r"C:\Users\A.W\Downloads\lookupdata.xlsx"
excel_sheet_name = "Store Name Data"
adjective_column_name = "Adjectives"
noun_column_name = "Nouns"


# fetch the sheet data in a dataframe
df = pd.read_excel(excel_file_path, sheet_name=excel_sheet_name)
# print(df)

# open the csv file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
# create the header
    header = ['StoreName','StoreType','StoreOpeningDate','Address','City','State','Region','ManagerName']
    
# write the header to the csv file
    writer.writerow(header)
    
# loop and generate multiple rows
    for _ in range(num_rows):
        
# select a random Adjective and Noun 
# and we are going to concatenate it with
# the word "The" and finally use that 
# as our store name
        random_adjective = df[adjective_column_name].sample(n=1).values[0]
        random_noun = df[noun_column_name].sample(n=1).values[0]
        store_name = f"The {random_adjective} {random_noun}"
#       print(store_name)

# Genearate a random row
        row = [
            store_name,
            random.choice(['Exclusive', 'MBO', 'SMB', 'Outlet Stores']),
            faker.date(),
            faker.address().replace("\n", " ").replace(",", ""),
            faker.city(),
            faker.state(),
            random.choice(['North', 'South', 'East', 'West']),
            faker.first_name()
        ]

# write the row to the csv file
        writer.writerow(row)

# print success statement
print("The process completed successfully.")
```
