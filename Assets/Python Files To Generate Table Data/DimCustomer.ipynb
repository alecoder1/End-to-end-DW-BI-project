```python
# import python libraries
import csv
import random
from faker import Faker

```

```python
# Initialize Faker
faker = Faker()

# Input the number of records that the csv should have
num_rows = int(input("Enter the number of rows the csv should have: "))
csv_file = input(" Enter the name of the csv file (with .csv extension): ")
    
# print success statement
print("The file has been loaded successfully")
```

```python
# import the csv file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
# create the header
    header = ['FirstName', 'LastName', 'Gender', 'DateOfBirth', 'Email', 'PhoneNumber', 'Address', 'City','State', 'PostalCode', 'Country', 'LoyaltyProgramID']
    
# Write the header to the CSV file
    writer.writerow(header)
    
# Loop and generate multiple rows of data
    for _ in range(num_rows):
        row = [
            faker.first_name(),
            faker.last_name(),
            random.choice(['M', 'F', 'Others', 'Not Specified']),
            faker.date_of_birth(minimum_age=18, maximum_age=90),
            faker.email(),
            faker.phone_number(),
            faker.address().replace(","," ").replace("\n"," "),
            faker.city(),
            faker.state(),
            faker.postcode(),
            faker.country(),
            random.randint(1, 5)
        ]
# Write the values into our csv file
        writer.writerow(row)

# print success statement
print("The file has been loaded successfully")
```
