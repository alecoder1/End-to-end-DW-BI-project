```python
import pyodbc

# Connect
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=DEC;DATABASE=ProjectDB', trusted_connection='yes')

# print("Connected with Windows Authentication")

cursor = conn.cursor()

# Execute a query
cursor.execute('SELECT * FROM DimLoyaltyProgram')

# Fetch all results
rows = cursor.fetchall()

# Display results
for row in rows:
    print(row)
```

```python
import pyodbc
import pandas as pd

# Connect
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=DEC;DATABASE=ProjectDB', trusted_connection='yes')

# Read entire data in a DataFrame
query = 'SELECT * FROM DimLoyaltyProgram'
df = pd.read_sql(query, conn)
print(df)
```

```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine

# Create SQLAlchemy engine
server = 'DEC'  # Update with your server name
database = 'ProjectDB'

# For Windows Authentication
engine = create_engine(f'mssql+pyodbc://{server}/{database}?driver=SQL+Server&trusted_connection=yes')

# Read data
query = 'SELECT * FROM DimLoyaltyProgram'
df = pd.read_sql(query, engine)

print(df)

df.to_csv('LoyaltyProgram.csv', index=False)
```
