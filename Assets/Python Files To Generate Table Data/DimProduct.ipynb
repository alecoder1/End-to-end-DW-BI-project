```python
import pandas as pd
import csv
import random

```

```python


# input the number of rows that the csv file should have
num_rows = int(input("Enter the number of rows you want to generate in te csv file: "))

# input the name of the csv file with the (.csv extension)
csv_file = input("Enter the name of the csv file: ")

# Details of the excel file that has the lookup data, 
# File Path and Name, Sheet Name 
# and column names where the data is present

excel_file_path = r"C:\Users\A.W\Downloads\lookupdata.xlsx"
excel_sheet_name_product = "Raw Product Names"
product_column_name = "Product Name"
excel_sheet_name_category = "Product Categories"
category_column_name = "Category Name"


# fetch the sheet data in a dataframe
df = pd.read_excel(excel_file_path, sheet_name=excel_sheet_name_product)
df_cat = pd.read_excel(excel_file_path, sheet_name=excel_sheet_name_category)
# print(df)

# open the csv file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
# create the header
    header = ['ProductName','Category','Brand','UnitPrice']
    
# write the header to the csv file
    writer.writerow(header)
    
# loop and generate multiple rows
    for _ in range(num_rows):
        


# Genearate a random row
        row = [
            df[product_column_name].sample(n=1).values[0], # Product Name from excel
            df_cat[category_column_name].sample(n=1).values[0],    # Category Name from excel
            random.choice(['FakeLuxeAura', 'FakeUrbanGlow', 'FakeEtherealEdge', 'FakeVelvetVista', 'FakZenithStyle']), # Brand
            random.randint(100, 1000)
            
        ]

# write the row to the csv file
        writer.writerow(row)

# print success statement
print("The process completed successfully.")
```
